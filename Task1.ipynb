```python
#Trích xuất các dữ liệu mẫu (x, y) trong đó x là câu nói (utterance) và y là dialogue act tương ứng của x:
import json

def extract_utterances_and_dialogue_acts(data):
    samples = []
    for dialogue in data:
      for turn in dialogue["turns"]:
          if turn["speaker"] == "USER":
            utterance = turn["utterance"]
            dialogue_acts = []
            if "frames" in turn:
                for frame in turn["frames"]:
                    if "state" in frame and "active_intent" in frame["state"] and frame["state"]["active_intent"] != "NONE":
                      service = frame["service"]
                      active_intent = frame["state"]["active_intent"]
                      dialogue_acts.append(f"{service}-{active_intent}")
            if dialogue_acts:
                 samples.append((utterance, dialogue_acts))

    return samples
path = 'E:\\NLP\\AS3\\multiwoz-master\\data\\MultiWOZ_2.2\\train\\dialogues_012.json'
with open(path, 'r', encoding='utf-8') as file: 
    data = json.load(file)
samples = extract_utterances_and_dialogue_acts(data)
```

```python
#In ra kiểm tra kết quảquả
for utterance, dialogue_act in samples:
            print(f"Utterance: {utterance}")
            print(f"Dialogue Act: {dialogue_act}")
            print("-" * 24) # cho đẹp và dễ nhìn
```
